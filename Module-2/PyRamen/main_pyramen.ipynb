```python
import csv

# set the source file path
csv_path_menu = '/Users/christiankunz/PycharmProjects/envs/dev/pythonProject/menu_data.csv'
csv_path_sales = '/Users/christiankunz/PycharmProjects/envs/dev/pythonProject/sales_data.csv'

# initiate lists/dicts
menu = []
sales = []

report = {}

# initiate variables
total_quantity_ramen_type = 0
total_revenue_ramen_type = 0
total_cogs_ramen_type = 0
total_profit_ramen_type = 0
menu_item = ""

# open csv file in read mode
with open(csv_path_menu, "r") as menu_data:
    csv_reader = csv.reader(menu_data)

    # this skips the first row of the CSV file
    next(csv_reader)

    # read the data and store each combination in dict
    for row in csv_reader:
        menu.append(row)

with open(csv_path_sales, "r") as sales_data:
    csv_reader = csv.reader(sales_data)

    # this skips the first row of the CSV file
    next(csv_reader)

    # read the data and store each combination in dict
    for row in csv_reader:
        sales.append(row)

# iterate over each imported list
for sales_row in sales:
    sales_item = sales_row[4]
    total_quantity_ramen_type = int(sales_row[3])

    for menu_row in menu:
        menu_item = menu_row[0]
        total_revenue_ramen_type = float(menu_row[3])
        total_cogs_ramen_type = float(menu_row[4])
        total_profit_ramen_type = float(total_revenue_ramen_type - total_cogs_ramen_type)

        # ramen type from sales list also needs to be in menu list. Other ramen types are not needed as it has not been sold.
        # if true - set report dict with initial combination otherwise update existing records
        if sales_item == menu_item:
            if sales_item not in report.keys():
                report[sales_item] = {"01-count": total_quantity_ramen_type, "02-revenue": total_revenue_ramen_type, "03-cogs": total_cogs_ramen_type, "04-profit": total_profit_ramen_type}
            else:
                report[sales_item]["01-count"] += total_quantity_ramen_type
                report[sales_item]["02-revenue"] += total_revenue_ramen_type * total_quantity_ramen_type
                report[sales_item]["03-cogs"] += total_cogs_ramen_type * total_quantity_ramen_type
                report[sales_item]["04-profit"] += total_profit_ramen_type * total_quantity_ramen_type

# set the path for pyramen_output_data.txt
output_path = "/Users/christiankunz/PycharmProjects/envs/dev/pythonProject/pyramen_output_data_2.txt"

with open(output_path, "w") as txt_file:
    print(report, file=txt_file)
```
